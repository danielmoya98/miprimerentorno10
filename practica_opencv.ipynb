```python
import cv2
import matplotlib.pyplot as plt

def mostrar_imagen(imagen, titulo="Imagen"): 
    imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    plt.imshow(imagen_rgb)
    plt.title(titulo)
    plt.axis('off') 
    plt.show()
    
imagen_bgr=cv2.imread('C:/Users/Alienware/Downloads/milei.jpg')
mostrar_imagen(imagen_bgr, "Imagen en formato BGR")
imagen_gris = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite('imagen_gris.jpg', imagen_gris)
mostrar_imagen(imagen_gris, "Imagen en escala de grises")
```
