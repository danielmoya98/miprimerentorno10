```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imagen_generada_gradiente_lineal():
    height = 300
    width = 300
    img = np.zeros((height, width), dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            img[i, j] = int(255 * (i + j) / (height + width))  # Gradiente lineal
    return img

def umbralizacion_isodata(img):
    _, thresholded = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return thresholded

def umbralizacion_picos(img):
    gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, np.ones((3,3), dtype=np.uint8))
    _, thresholded = cv2.threshold(gradient, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return thresholded

def umbralizacion_adaptativa_media(img):
    thresholded = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    return thresholded

def umbralizacion_adaptativa_gaussiano(img):
    thresholded = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    return thresholded

def umbralizacion_adaptativa_binario(img):
    thresholded = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    return thresholded

def umbralizacion_adaptativa_binario_inverso(img):
    thresholded = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    return thresholded

def umbralizacion_variable(img):
    _, thresholded = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return thresholded

def umbralizacion_trunc(img):
    _, thresholded = cv2.threshold(img, 127, 200, cv2.THRESH_TRUNC)
    return thresholded

def umbralizacion_tozero(img):
    _, thresholded = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
    return thresholded

def umbralizacion_tozero_inv(img):
    _, thresholded = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
    return thresholded

def umbralizacion_trunc_inv(img):
    _, thresholded = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
    return 255 - thresholded

def mostrar_imagen(imagen, titulo):
    plt.figure(figsize=(6, 6))
    plt.imshow(imagen, cmap='gray')
    plt.title(titulo)
    plt.axis('off')
    plt.show()

# Generar la imagen
imagen_gradiente_lineal = imagen_generada_gradiente_lineal()

# Mostrar la imagen original
mostrar_imagen(imagen_gradiente_lineal, "Imagen Gradiente Lineal")

# Aplicar umbralización ISODATA
umb_isodata = umbralizacion_isodata(imagen_gradiente_lineal)
mostrar_imagen(umb_isodata, "Umbralizacion ISODATA")

# Aplicar umbralización de picos
umb_picos = umbralizacion_picos(imagen_gradiente_lineal)
mostrar_imagen(umb_picos, "Umbralizacion de Picos")

# Aplicar umbralización adaptativa (media)
umb_adaptativa_media = umbralizacion_adaptativa_media(imagen_gradiente_lineal)
mostrar_imagen(umb_adaptativa_media, "Umbralizacion Adaptativa (Media)")

# Aplicar umbralización adaptativa (gaussiano)
umb_adaptativa_gaussiano = umbralizacion_adaptativa_gaussiano(imagen_gradiente_lineal)
mostrar_imagen(umb_adaptativa_gaussiano, "Umbralizacion Adaptativa (Gaussiano)")

# Aplicar umbralización adaptativa (binario)
umb_adaptativa_binario = umbralizacion_adaptativa_binario(imagen_gradiente_lineal)
mostrar_imagen(umb_adaptativa_binario, "Umbralizacion Adaptativa (Binario)")

# Aplicar umbralización adaptativa (binario inverso)
umb_adaptativa_binario_inverso = umbralizacion_adaptativa_binario_inverso(imagen_gradiente_lineal)
mostrar_imagen(umb_adaptativa_binario_inverso, "Umbralizacion Adaptativa (Binario Inverso)")

# Aplicar umbralización variable
umb_variable = umbralizacion_variable(imagen_gradiente_lineal)
mostrar_imagen(umb_variable, "Umbralizacion Variable")

# Aplicar umbralización Trunc
umb_trunc = umbralizacion_trunc(imagen_gradiente_lineal)
mostrar_imagen(umb_trunc, "Umbralizacion Trunc")

# Aplicar umbralización Trunc Inversa
umb_trunc_inv = umbralizacion_trunc_inv(imagen_gradiente_lineal)
mostrar_imagen(umb_trunc_inv, "Umbralizacion Trunc Inversa")


# Aplicar umbralización Tozero
umb_tozero = umbralizacion_tozero(imagen_gradiente_lineal)
mostrar_imagen(umb_tozero, "Umbralizacion Tozero")

# Aplicar umbralización Tozero Inverso
umb_tozero_inv = umbralizacion_tozero_inv(imagen_gradiente_lineal)
mostrar_imagen(umb_tozero_inv, "Umbralizacion Tozero Inverso")
```
