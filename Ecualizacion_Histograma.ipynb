```python
##algoritmo de cv2

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carga de la imagen en escala de grises
img = cv2.imread('C:/Users/Alienware/Downloads/libertad.jpg', cv2.IMREAD_GRAYSCALE)

# Ecualizacin del histograma
img_ecualizada = cv2.equalizeHist(img)

# Calcular histogramas
histograma = cv2.calcHist([img], [0], None, [256], [0, 256])
histograma_ecualizado = cv2.calcHist([img_ecualizada], [0], None, [256], [0, 256])

# Mostrar las imgenes y sus histogramas
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 2, 3)
plt.plot(histograma, color='b')
plt.title('Histograma Original')

plt.subplot(2, 2, 4)
plt.plot(histograma_ecualizado, color='r')
plt.title('Histograma Ecualizado')

plt.tight_layout()
plt.show()
```

```python
##algoritmo de kleif

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen en escala de grises
img = cv2.imread('C:/Users/Alienware/Downloads/libertad.jpg', cv2.IMREAD_GRAYSCALE)

# Crear el objeto CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Aplicar CLAHE a la imagen
img_nueva = clahe.apply(img)

# Calcular histogramas
histograma = cv2.calcHist([img], [0], None, [256], [0, 256])
histograma_nueva = cv2.calcHist([img_nueva], [0], None, [256], [0, 256])

# Mostrar las imgenes y sus histogramas
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.imshow(img_nueva, cmap='gray')
plt.title('Imagen Nueva')

plt.subplot(2, 2, 3)
plt.plot(histograma, color='b')
plt.title('Histograma Original')

plt.subplot(2, 2, 4)
plt.plot(histograma_nueva, color='r')
plt.title('Histograma con CLAHE')

plt.tight_layout()
plt.show()


```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en color
img = cv2.imread('C:/Users/Alienware/Downloads/libertad.jpg')

# Convertir la imagen a espacio de color HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Separar los canales de color
h, s, v = cv2.split(img_hsv)

# Ecualizar los canales por separado
v_ecualizada = cv2.equalizeHist(v)
s_ecualizada = cv2.equalizeHist(s)
h_ecualizada = cv2.equalizeHist(h)

# Fusionar los canales de nuevo en la imagen HSV ecualizada
img_hsv_ecualizada = cv2.merge([h_ecualizada, s_ecualizada, v_ecualizada])

# Convertir la imagen ecualizada de nuevo a espacio de color BGR
img_ecualizada = cv2.cvtColor(img_hsv_ecualizada, cv2.COLOR_HSV2BGR)

# Calcular los histogramas de los canales ecualizados
hist_h = cv2.calcHist([h_ecualizada], [0], None, [256], [0, 256])
hist_s = cv2.calcHist([s_ecualizada], [0], None, [256], [0, 256])
hist_v = cv2.calcHist([v_ecualizada], [0], None, [256], [0, 256])

# Mostrar la imagen original y su histograma
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.plot(hist_h, color='r')
plt.title('Histograma de H')

plt.subplot(2, 2, 3)
plt.plot(hist_s, color='g')
plt.title('Histograma de S')

plt.subplot(2, 2, 4)
plt.plot(hist_v, color='b')
plt.title('Histograma de V')

plt.tight_layout()
plt.show()

# Mostrar la imagen ecualizada fusionada y su histograma
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img_ecualizada, cv2.COLOR_BGR2RGB))
plt.title('Imagen Ecualizada Fusionada')

hist_ecualizada = cv2.calcHist([img_hsv_ecualizada], [0], None, [256], [0, 256])

plt.subplot(2, 2, 2)
plt.plot(hist_ecualizada, color='gray')
plt.title('Histograma Ecualizado')

plt.tight_layout()
plt.show()



plt.show()

```
