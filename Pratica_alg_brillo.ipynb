```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = cv2.imread('C:/Users/Alienware/Downloads/daniel.png')

# Ajuste de Valor
brillo_alto = cv2.add(imagen, 50)  # Aumentar el brillo
brillo_bajo = cv2.subtract(imagen, 50)  # Disminuir el brillo

# Ajuste Gamma
gamma = 1.5
ajuste_gamma = np.array(255*(imagen / 255) ** gamma, dtype='uint8')

# Histograma de Equalización
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
hist_eq = cv2.equalizeHist(imagen_gris)

# Filtrado Bilateral
filtrado_bilateral = cv2.bilateralFilter(imagen, 9, 75, 75)

# Mostrar las imágenes
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(imagen)
plt.title('Daniel Moya')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(brillo_alto, cv2.COLOR_BGR2RGB))
plt.title('Ajuste de Valor (Alto Brillo)')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(brillo_bajo, cv2.COLOR_BGR2RGB))
plt.title('Ajuste de Valor (Bajo Brillo)')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(ajuste_gamma, cv2.COLOR_BGR2RGB))
plt.title('Ajuste Gamma')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(hist_eq, cmap='gray')
plt.title('Histograma de Equalización')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(cv2.cvtColor(filtrado_bilateral, cv2.COLOR_BGR2RGB))
plt.title('Filtrado Bilateral')
plt.axis('off')

plt.show()
```
