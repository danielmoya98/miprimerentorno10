```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

paletas = [
    "COLORMAP_AUTUMN",
    "COLORMAP_BONE",
    "COLORMAP_JET",
    "COLORMAP_WINTER",
    "COLORMAP_RAINBOW",
    "COLORMAP_OCEAN",
    "COLORMAP_SUMMER",
    "COLORMAP_SPRING",
    "COLORMAP_COOL",
    "COLORMAP_HSV",
    "COLORMAP_PINK",
    "COLORMAP_HOT",
    "COLORMAP_PARULA",
    "COLORMAP_MAGMA",
    "COLORMAP_INFERNO",
    "COLORMAP_PLASMA",
    "COLORMAP_VIRIDIS",
    "COLORMAP_CIVIDIS",
    "COLORMAP_TWILIGHT",
    "COLORMAP_TWILIGHT_SHIFTED",
    "COLORMAP_TURBO",
    "COLORMAP_DEEPGREEN"
]

img_original = cv2.imread("C:/Users/Alienware/Downloads/libertad.jpg")

total_imagenes = len(paletas) + 1

num_filas = total_imagenes // 5 + (total_imagenes % 5 > 0)
num_columnas = min(total_imagenes, 5)

fig, axes = plt.subplots(num_filas, num_columnas, figsize=(3*num_columnas, 3*num_filas))

axes[0, 0].imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title('Imagen Original')

for idx, palette in enumerate(paletas):
    row = (idx + 1) // num_columnas
    col = (idx + 1) % num_columnas
    colormap = getattr(cv2, palette)
    color_img = cv2.applyColorMap(img_original, colormap)
    axes[row, col].imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
    axes[row, col].set_title('Paleta: ' + palette)

for ax in axes.flatten():
    ax.axis('off')

plt.tight_layout()
plt.show()
```
